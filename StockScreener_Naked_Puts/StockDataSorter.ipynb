```python
def is_market_closed(target_date):
    # Define U.S. stock market hours (Eastern Time)
    market_close_time = time(16, 0)
    
    # Define Eastern Time Zone
    eastern = timezone('US/Eastern')
    
    # Get the current time in Eastern Time Zone
    current_time = datetime.now(eastern)
    
    # If the target_date is in the future, the market is definitely not closed
    if target_date.date() > current_time.date():
        return False
    
    # If the target_date is a weekend, the market is closed
    if target_date.weekday() >= 5:
        return True
    
    # For holidays, you would normally have a list of market holidays and check here
    # Skipping that part for simplicity
    
    # If the current time is past the market close time and the date is today or in the past, market is closed
    if target_date.date() == current_time.date():
        return current_time.time() >= market_close_time
    
    # For all past dates that are not weekends or holidays, the market is closed
    return True
```

```python
def get_batch_ticker_prices(tickers, date):
    target_date = datetime.strptime(date, '%Y-%m-%d')
    
    if not is_market_closed(target_date):
        print("The market has not closed yet for this date.")
        return
    
    # Your existing code to fetch data
    data = yf.download(tickers, start=date, end=date)
    
    if isinstance(data.columns, pd.MultiIndex):
        close_prices = data['Close']
    else:
        close_prices = data['Close']
        
    return close_prices.to_dict()
```
